```python
import pandas as pd

# Load CSV files
df1 = pd.read_csv('db_building_A.csv')
df2 = pd.read_csv('db_building_B.csv')

# Perform natural join on the common column, e.g., 'DATE'
result = pd.merge(df1, df2, on='DATE')

# Save or display the result
result.to_csv('joined.csv', index=False)
print(result)
```

```python
import pandas as pd

# Load CSV files
df1 = pd.read_csv('db_building_A.csv')
df2 = pd.read_csv('db_building_B.csv')

# Perform natural join on all common columns
common_columns = list(set(df1.columns) & set(df2.columns))
result = pd.merge(df1, df2, on=common_columns)

# Save or display the result
result.to_csv('joined1.csv', index=False)
print(result)
```
