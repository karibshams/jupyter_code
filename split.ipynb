```python
import os
import random
import shutil

# Define the source and destination directories
source_dir = r"D:\CAPSTONE 400A\final_dataset\Tuberculosis\TB_Chest_Radiography_Database\convertdcmtopng"
destination_dir = r"D:\CAPSTONE 400A\final_dataset\Tuberculosis\TB_Chest_Radiography_Database\image7500f\Tuberculosis"

# Make sure the destination directory exists
os.makedirs(destination_dir, exist_ok=True)

# Get a list of all files in the source directory
all_files = os.listdir(source_dir)

# Ensure the number of files in the source directory is as expected
if len(all_files) != 7316:
    raise ValueError(f"Expected 7316 files, but found {len(all_files)} in the source directory.")

# Randomly select 3500 files
selected_files = random.sample(all_files, 3500)

# Copy the selected files to the destination directory
for file_name in selected_files:
    src_file = os.path.join(source_dir, file_name)
    dst_file = os.path.join(destination_dir, file_name)
    shutil.copy(src_file, dst_file)

print(f"Successfully copied 3500 images to {destination_dir}")
```

```python
import os
import pydicom
from PIL import Image
import numpy as np

# Specify the directory containing DICOM files and the output directory
input_dir = r"H:\TB_Portals_Published_CXRs_August_2023"
output_dir = r"D:\CAPSTONE 400A\final_dataset\Tuberculosis\TB_Chest_Radiography_Database\convertdcmtopng"


# Create the output directory if it does not exist
os.makedirs(output_dir, exist_ok=True)

def convert_dcm_to_png(dcm_path, png_path, size=(224, 224)):
    try:
        # Read the DICOM file
        dicom = pydicom.dcmread(dcm_path)
        # Get the pixel array from the DICOM file
        pixel_array = dicom.pixel_array
        # Convert the pixel array to an 8-bit image (0-255)
        image = Image.fromarray(np.uint8((pixel_array / np.max(pixel_array)) * 255))
        # Resize the image to the desired size
        image = image.resize(size, Image.Resampling.LANCZOS)
        # Save the image as a PNG file
        image.save(png_path)
    except Exception as e:
        print(f"Error converting {dcm_path}: {e}")

# Traverse the input directory and convert all DICOM files to PNG
for root, _, files in os.walk(input_dir):
    for file in files:
        if file.lower().endswith(".dcm"):
            dcm_path = os.path.join(root, file)
            png_filename = os.path.splitext(file)[0] + ".png"
            png_path = os.path.join(output_dir, png_filename)
            if os.path.exists(dcm_path):
                convert_dcm_to_png(dcm_path, png_path)
                print(f"Converted {dcm_path} to {png_path}")
            else:
                print(f"File not found: {dcm_path}")

print("Conversion complete.")
```

```python
import os
import pandas as pd

# Define the directory containing the images
data_dir = "D:/CAPSTONE 400A/final_dataset/Tuberculosis/TB_Chest_Radiography_Database/New_folder"

# Get the list of image filenames for each class
normal_images = [os.path.join(data_dir, "Normal", filename) for filename in os.listdir(os.path.join(data_dir, "Normal"))]
tb_images = [os.path.join(data_dir, "Tuberculosis", filename) for filename in os.listdir(os.path.join(data_dir, "Tuberculosis"))]

# Create DataFrame to store information
df_normal = pd.DataFrame({'filename': normal_images, 'class': 'Normal'})
df_tb = pd.DataFrame({'filename': tb_images, 'class': 'Tuberculosis'})
df = pd.concat([df_normal, df_tb], ignore_index=True)

# Save the DataFrame to a CSV file
csv_file_path = os.path.join(data_dir, 'image_data.csv')
df.to_csv(csv_file_path, index=False)

print("CSV file saved successfully at:", csv_file_path)
```
